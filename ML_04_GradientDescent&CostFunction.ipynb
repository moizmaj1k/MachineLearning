```python
import numpy as np
import math

'''

Implementation of Gradient Descent
Consider X = [1, 2, 3, 4, 5]    --->    Features
and,     y = [5, 7, 9, 11, 13]  --->    Targets

'''

def gradient_descent(x,y):
  # Setting initial values for slope and intercept
  m_curr = b_curr = 0
  iterations = 1000000 # Setting iterations, we will fine tune them later
  n = len(x) # total number of data points
  learning_rate = 0.0002 # Learning rate can be adjusted later on
  cost_prev = 0

  for i in range(iterations):
    y_pred = m_curr * x + b_curr
    # Implementing derivative equations
    mDerivative = -(2/n)*sum(x*(y-y_pred))
    bDerivative = -(2/n)*sum(y-y_pred)
    # Now, we write out step equations
    m_curr = m_curr - learning_rate*mDerivative
    b_curr = b_curr - learning_rate*bDerivative
    # Calculating the cost in each iteration
    new_cost = (1/n) * sum([val**2 for val in (y-y_pred)])
    if math.isclose(new_cost, cost_prev, rel_tol=1e-20):
      break
    cost_prev = new_cost
    print(f'm: {m_curr}  b: {b_curr}  cost(MSE): {new_cost}  iteration: {i}')




x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```

<h3>Required slope(m) was 2 and y-intercept(b) was 3</h3>

We are very close to our requirement and the MSE(Cost) is also less enough to consider this as the best fit line.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

import warnings
from sklearn.exceptions import ConvergenceWarning
with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=ConvergenceWarning)

df = pd.read_csv('ML_04_test_scores.csv')
df
```

```python
# In the given csv file. 'math' is our feature and 'cs' is our target variable
# We dont really need name as it doesnt give any value/correlation to target

df = df.drop(['name'], axis=1)
```

```python
# Lets plot this data to see the distribution

import matplotlib.pyplot as plt
%matplotlib inline

# Scatter plot
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df['math'], df['cs'], color='red', marker='+')
```

```python
# Now we will be using our gradient descent algorithm on this data

x = np.array(df['math'])
y = np.array(df['cs'])

gradient_descent(x,y)
```

```python
# Implementing sklearn LinearRegression to see the values of m and b. So we can compare them

lr = LinearRegression()
lr.fit(df[['math']], df[['cs']])

print(f'm(slope): ', lr.coef_)
print(f'b(y-intercecpt): ', lr.intercept_)
```

comparing the results of gradient_descent algorithm and the sklearn LinearRegression() mode we can see that we managed to get the same results for 'm' and 'b' using the appropriate vaules of alpha(learning rate) and number of iterations.
