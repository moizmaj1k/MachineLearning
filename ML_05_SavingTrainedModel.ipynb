```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
import warnings
from sklearn.exceptions import ConvergenceWarning
with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
df = pd.read_csv('ML_02_house_price.csv')
df.head()
```

```python
df = df.drop('rooms', axis=1)
df
```

```python
# Fitting a linear regression model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(df[['area']], df[['price']])
```

```python
lr.predict([[5000]])
```

<h3>We did this in ML_02 lecture. This is a trained linear regression model and we will now learn how to save it for using the same trained model later on without needing to train it again.</h3>

<h3>Approach 1(pickle library)</h3>

```python
import pickle
# pickle is a library for dumping and loading files
```

```python
with open('model_pickle', 'wb') as f:
  pickle.dump(lr, f)

# This will create a file named 'model_pickle'
# wb(write binary) we will write binary data to this file
# pickle.dump will take our model and save it in the file
```

```python
with open('model_pickle', 'rb') as f:
  model = pickle.load(f)

# This opens the file we saved in read binary mode
# we can load the model in a variable and this way we have our pretrained model
```

```python
model.predict([[5000]])
```
